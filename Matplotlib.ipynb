```python
import pandas as pd
```

```python
fbk = ['Facebook', 2449, True, 2006]
twt = ['Twitter', 339, False, 2006]
ig = ['Instagram', 1000, True, 2010]
yt = ['YouTube', 2000, False, 2005]
lkn = ['LinkedIn', 663, False, 2003]
wsp = ['WhatsApp', 1600, True, 2009]

lista_rrss = [fbk, twt, ig, yt, lkn, wsp]

#crear dataframe a partir de listas
df_rrss = pd.DataFrame(lista_rrss, columns = ['Nombre', 'Cantidad', 'ES_FBK', 'Año'])
df_rrss
```

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,8,27,64,125]

plt.plot(x,y)
plt.show()
```

```python
# gráfico de líneas utilizando matplotlib
plt.plot(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
# gráfico de puntos utilizando matplotlib
plt.scatter(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
# EXTRA: cómo hacer lo mismo con métodos de un dataframe de pandas
df_rrss.plot(kind='scatter', x='Nombre', y='Cantidad')
plt.show()
```

```python
# gráficos de barras utilizando matplotlib
plt.bar(df_rrss['Nombre'], df_rrss['Cantidad'])
plt.show()
```

```python
#gráficos de barras ordenadas descendentemente
df_rrss_sort = df_rrss.sort_values('Cantidad', ascending=False)
plt.bar(df_rrss_sort['Nombre'], df_rrss_sort['Cantidad'])
plt.show()
```

```python
#gráfico de barras ordenado con color

df_rrss_sort = df_rrss.sort_values('Cantidad', ascending=False)
plt.bar(df_rrss_sort['Nombre'], df_rrss_sort['Cantidad'], color = ['b','r','g','m','k','c'])
plt.show()

# b : blue,
# g : green,
# r : red,
# c : cyan,
# m : magenta,
# y : yellow,
# b : black,
# w : white,
```

```python
#gráfico de pie utilizando matplotlib

plt.pie(df_rrss['Cantidad'], labels = df_rrss['Nombre'])
plt.show()
```

```python
#gráfico de puntos utilizando matplotlib
lista_col_ress = ['#3b5998', '#3b5238', '#3b1098', '#3b9098', '#3b5098']

plt.pie(df_rrss_sort['Cantidad'], labels = df_rrss_sort['Nombre'], colors = lista_col_ress)
plt.show()
```
